```python
import pandas as pd
import matplotlib as plt
import numpy as np
%matplotlib inline
```

```python
data = pd.read_csv('valorant_weapons.csv')
```

```python
data.head()
```

```python
data['cadence_shooting']=data['cadence_shooting'].astype('str').str.replace(',','.')
data['cadence_shooting']=pd.to_numeric(data['cadence_shooting'], errors='ignore')
```

```python
def changer(column):
    data[column]=data[column].astype('str').str.replace(',','.')
    data[column]=pd.to_numeric(data[column], errors='ignore')
    
changer('velocity_to_equip')
changer('dispersion_bullet')
changer('dispersion_bullet_scope')
changer('velocity_recharge')
```

```python
SPlow = data[:][data['shoting_penetretion'] == 'low']
SPlow
```

```python
SPmed = data[:][data['shoting_penetretion'] == 'medium']
SPmed
```

```python
SPhigh = data[:][data['shoting_penetretion'] == 'high']
SPhigh
```
